```python
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}
print("Set A:", A)
print("Set B:", B)
print("Union (A | B):", A | B)
print("Union using method:", A.union(B))
print("Intersection (A & B):", A & B)
print("Intersection using method:", A.intersection(B))
print("Difference (A - B):", A - B)
print("Difference (B - A):", B - A)
print("Symmetric Difference (A ^ B):", A ^ B)
print("Symmetric Difference using method:", A.symmetric_difference(B))
C = {1, 2, 3}
print("Is C subset of A?", C.issubset(A))
print("Is A superset of C?", A.issuperset(C))
A.add(10)
print("After adding 10 to A:", A)
A.remove(2)
print("After removing 2 from A:", A)
A.discard(100)
print("After discard 100:", A)
B.clear()
print("After clearing B:", B)
```
