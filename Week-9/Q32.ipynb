```python
#32)Implement the lasso model
from sklearn.linear_model import Lasso
import numpy as np
X=np.array([[1,25,160],[2,30,165],[3,35,170]])
Y=np.array([30000,40000,50000])
model=Lasso(alpha=0.1)
model.fit(X,Y)
print(model.coef_)
```
