```python
#33)Implement the similarity and distances for any dataset
from scipy.spatial import distance
from scipy.stats import  pearsonr
import numpy as np
p1=np.array([2,3])
p2=np.array([5,7])
euc=distance.euclidean(p1,p2)
mh=distance.cityblock(p1,p2)
cb=distance.chebyshev(p1,p2)
hm=distance.cosine(p1,p2)
cs=1-cb
set1,set2=set(p1),set(p2)
js=len(set1&set2)|len(set1&set2)
r=pearsonr(p1,p2)
print("Euclidean Distance",euc)
print("Manhatten Distance",mh)
print("Chebyshev Distance",cb)
print("Cosine Similarity",cs)
print("Jaccord Distance",js)
print("Pearsorns coefficient",r)
```
