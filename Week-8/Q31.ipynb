```python
#31)Implement the random forest for the given dataset
from sklearn.ensemble import RandomForestClassifier
X=[[1,1],[1,0],[0,1],[0,0]]
y=[0,1,1,0]
model=RandomForestClassifier(n_estimators=100,random_state=42)
model.fit(X,y)
print(model.predict([[1,0]]))
```
