```python
#28)Write the code to implement Support vector machines(SVM)
from sklearn import svm
X=[[2,2],[4,4],[2,1],[0,2]]
Y=[1,1,-1,-1]
model=svm.SVC(kernel='linear')
model.fit(X,Y)
print(model.predict([[0,0]]))
```
