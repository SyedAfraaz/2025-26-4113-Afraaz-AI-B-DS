```python
#question 13-- eda operations
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
data=[10,12,15,18,20,22,25,30,35,40]
df=pd.DataFrame(data,columns=['values'])
print("Dataset:")
print(df)
print("\nSummary statistics:")
print(df.describe())
print("\nVariance:",np.var(data))
print("\nStandard deviation:",np.std(data))
plt.hist(data)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Data Distribution')
plt.show()
mean_value=np.mean(data)
median_value=np.median(data)
mode_value = df['values'].mode()
print("\nInference:")
print("Mean:",mean_value)
print("Median:",median_value)
if len(mode_value) == 1:
    print("Mode:", mode_value[0])
else:
    print("Mode:", list(mode_value))
if mean_value>median_value:
    print("The distribution is positively skewed.")
elif mean_value<median_value:
    print("The distribution is negatively skewed.")
else:
    print("The distribution is symmetric.")
```

```python
#question 14-- eda operations in dataset

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
print("Dataset:")
df = pd.read_csv(url)
print(df.head())
data = df['sepal_length']
print("\nSummary statistics:")
print(data.describe())
print("\nVariance:", np.var(data))
print("Standard deviation:", np.std(data))
# Mean, Median, Mode
mean_value = np.mean(data)
median_value = np.median(data)
mode_value = data.mode()
# Histogram
plt.hist(data, bins=10, color='skyblue', edgecolor='black')
plt.axvline(mean_value, color='red', linestyle='dashed',
            linewidth=2, label=f"Mean: {mean_value:.2f}")
plt.axvline(median_value, color='green', linestyle='dashed',
            linewidth=2, label=f"Median: {median_value:.2f}")
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.title('Distribution of Sepal Length')
plt.legend()
plt.show()
print("\nInference:")
print("Mean:", mean_value)
print("Median:", median_value)
if len(mode_value) == 1:
    print("Mode:", mode_value[0])
else:
    print("Mode:", list(mode_value))
if mean_value > median_value:
    print("The distribution is positively skewed.")
elif mean_value < median_value:
    print("The distribution is negatively skewed.")
else:
    print("The distribution is symmetric.")
```
