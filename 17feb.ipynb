```python
#Performing data manipulations, data operations, ranking on a dataset
import pandas as pd
url="https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df=pd.read_csv(url)
print("Original Dataset:")
print(df.head())
print("\nRecords where Total_Bill>20 and Tips>3")
filtered_and=df[(df['total_bill']>20)&(df['tip']>3)]
print(filtered_and)
print("\nRecords where Day is Saturday or Sunday")
filtered_or=df[(df['day']=='Sat')|(df['day']=='Sun')]
print(filtered_or)
print("\nRecords where Smoker is NOR Yes")
filtered_not=df[~(df['smoker']=='Yes')]
print(filtered_not)
print("\nFemale and Non-Smoker and Total_Bill>15")
combined_filter=df[(df['sex']=='Female')&(df['smoker']=='No')&(df['total_bill']>15)]
print(combined_filter)
print("\nNumber of records satisfying combined condition:",len(combined_filter))
print("\nSorted by Total_Bill in ascending order:")
sorted_asc=df.sort_values(by='total_bill')
print(sorted_asc.head())
print("\nSorted by Tip in descending order:")
sorted_desc=df.sort_values(by='tip',ascending=False)
print(sorted_desc.head())
print("\nSorted by Day and Total_Bill")
sorted_multi=df.sort_values(by=['day','total_bill'])
print(sorted_multi.head())
print("\nAverage Total Bill by Day")
group_day=df.groupby('day')['total_bill'].mean()
print(group_day)
print("\nAverage Tip by Sex and Smoker")
group_multi=df.groupby(['sex','smoker'])['tip'].mean()
print(group_multi)
print("\nCount of Records by Day")
group_count=df.groupby('day').size()
print(group_count)
```

```python
# Performing data manipulations + visualization

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)

print("Original Dataset:")
print(df.head())

filtered_and = df[(df['total_bill'] > 20) & (df['tip'] > 3)]
filtered_or = df[(df['day'] == 'Sat') | (df['day'] == 'Sun')]
filtered_not = df[~(df['smoker'] == 'Yes')]
combined_filter = df[(df['sex'] == 'Female') &
                     (df['smoker'] == 'No') &
                     (df['total_bill'] > 15)]

sorted_asc = df.sort_values(by='total_bill')
sorted_desc = df.sort_values(by='tip', ascending=False)
sorted_multi = df.sort_values(by=['day', 'total_bill'])

group_day = df.groupby('day')['total_bill'].mean()
group_multi = df.groupby(['sex', 'smoker'])['tip'].mean()
group_count = df.groupby('day').size()

sns.set(style="whitegrid")

# Scatter Plot: Total Bill vs Tip
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='total_bill', y='tip', hue='sex')
plt.title("Total Bill vs Tip")
plt.show()

#  Bar Plot: Average Total Bill by Day
plt.figure(figsize=(8,5))
sns.barplot(x=group_day.index, y=group_day.values)
plt.title("Average Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Average Total Bill")
plt.show()

# Bar Plot: Average Tip by Sex and Smoker
plt.figure(figsize=(8,5))
group_multi.unstack().plot(kind='bar')
plt.title("Average Tip by Sex and Smoker")
plt.ylabel("Average Tip")
plt.xticks(rotation=0)
plt.show()

# Count Plot: Number of Records by Day
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='day')
plt.title("Count of Records by Day")
plt.show()

# Histogram: Distribution of Total Bill
plt.figure(figsize=(8,5))
sns.histplot(df['total_bill'], bins=20, kde=True)
plt.title("Distribution of Total Bill")
plt.show()

# Boxplot: Total Bill by Day
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='day', y='total_bill')
plt.title("Total Bill Distribution by Day")
plt.show()

# Boxplot: Tips for Filtered Condition
plt.figure(figsize=(8,5))
sns.boxplot(data=combined_filter, x='sex', y='tip')
plt.title("Tips (Female, Non-Smoker, Total Bill > 15)")
plt.show()
```

```python
#exploring data analysis(EDA) on simple data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data=[10,20,30,22,34,56,67,34,45,40]
df=pd.DataFrame(data,columns=['Age'])
print("Original Dataset:")
print(df)
print("\nSummery statistics:")
print(df.describe())
print("\nVariance:", np.var(data))
print("\nStandard Deviation:", np.std(data))
plt.hist(data)
plt.title("Data Distribution")
plt.ylabel("Frequency")
plt.show()
mean_value=np.mean(data)
median_value=np.median(data)
print("\nMean:", mean_value)
print("\nMedian:", median_value)
if mean_value>median_value:
  print("\ndata is slightly right-skewed")
elif mean_value<median_value:
  print("\ndata is slightly left-skewed")
else:
  print("\nData is symmetric")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
url = "https://raw.githubusercontent.com/GeostatsGuy/GeoDataSets/master/12_sample_data.csv"
df = pd.read_csv(url)
print("Dataset (first 10 rows):")
print(df.head(10))
data = df['Porosity']
print("\nSummary statistics:")
print(data.describe())
print("\nVariance:", np.var(data))
print("Standard deviation:", np.std(data))
mean_value = np.mean(data)
median_value = np.median(data)
plt.hist(data, bins=10, color="skyblue", edgecolor="black")
plt.axvline(mean_value, color="red", linestyle="dashed", linewidth=2,
            label=f'Mean = {mean_value:.2f}')
plt.axvline(median_value, color="green", linestyle="dashed", linewidth=2,
            label=f'Median = {median_value:.2f}')
plt.xlabel("Porosity (cm)")
plt.ylabel("Frequency")
plt.title("Distribution of Porosity")
plt.legend()
plt.show()
print("\nInference")
print("Mean value:", mean_value)
print("Median value:", median_value)
if mean_value > median_value:
    print("Data is slightly right-skewed")
elif mean_value < median_value:
    print("Data is slightly left-skewed")
else:
    print("Data is symmetric")


```
