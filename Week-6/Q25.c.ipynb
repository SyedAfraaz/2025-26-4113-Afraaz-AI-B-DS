```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression

# Sample dataset
data = pd.DataFrame({
    'review': [
        "I love this product", "This is amazing",
        "Worst experience ever", "I hate it",
        "Very good", "Not good"
    ],
    'sentiment': [1, 1, 0, 0, 1, 0]  # 1 = Positive
})

# Convert text
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(data['review'])
y = data['sentiment']

# Train
model = LogisticRegression()
model.fit(X, y)

# Test
review = ["I really love it"]
review_vec = vectorizer.transform(review)

print("Sentiment (1=Positive, 0=Negative):", model.predict(review_vec)[0])
```
